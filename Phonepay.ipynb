## 3.1: Ensure data consistency across state and district levels
## For each state, calculate the total number of transactions, total transaction amount, and total registered users by summing up the values from the district level data.

```python
dt=District_Txn.groupby(['State']).agg(Transactions=('Transactions','sum'),Transactionsamt=('Amount (INR)','sum'),RegisteredUsers=('Registered Users','sum'))
```

```python
st=State_TxnUsers.groupby(['State']).agg(Transactions=('Transactions','sum'),Transactionsamt=('Amount (INR)','sum'),RegisteredUsers=('Registered Users','sum'))
```

dt

```python
merged_df=pd.merge(dt,st,on='State')
```

## 2 Compare the results with the corresponding values at the state level to ensure they match.

```python
merged_df.loc[(merged_df['Transactionsamt_x'])!=(merged_df['Transactionsamt_y'])]
```

```python
import pandas as pd

def check_matching_values(df, columns):

    results = {}
    for col in columns:
        # Check if the values in the columns with suffix _x and _y are equal
        results[col] = (df[f'{col}_x'] == df[f'{col}_y']).all()
    
    # Convert the results to a DataFrame
    results_df = pd.DataFrame(list(results.items()), columns=['Column', 'All Values Match'])
    return results_df

# Assuming the merged DataFrame is named 'merged_df'
columns_to_check = ['Transactions', 'Transactionsamt', 'RegisteredUsers']
result_df = check_matching_values(merged_df, columns_to_check)

# Display the results
print(result_df)
```

## 5.1: Plot the total transactions and amount over time for a selected state

## 1 Create a line plot showing the total number of transactions and the total transaction amount over time (years and quarters) for a selected state.

```python
x=State_TxnSplit.loc[State_TxnSplit['State']=='Andaman & Nicobar Islands'].groupby(['Year','Quarter']).agg(Transactions=('Transactions','sum'), Amount_INR=('Amount (INR)','sum'))
```

```python
x.plot()
```

```python
# Reset the index to make 'Year' and 'Quarter' columns
x_reset = x.reset_index()

# Combine 'Year' and 'Quarter' to create a time-like index
x_reset['Year_Quarter'] = x_reset['Year'].astype(str) + ' Q' + x_reset['Quarter'].astype(str)

# Set the new 'Year_Quarter' column as the index
x_reset.set_index('Year_Quarter', inplace=True)

# Plot 'Transactions' over time
plt.figure(figsize=(10, 6))
plt.plot(x_reset.index, x_reset['Transactions'], marker='o')
plt.xlabel("Year and Quarter")
plt.ylabel("Transactions")
plt.title("Transactions over Time for Andaman & Nicobar Islands")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.plot(x_reset.index,x_reset['Amount_INR'],marker='o')
plt.xlabel("Year and Quarter")
plt.ylabel("Amount")
plt.title("Amount over Time for Andaman & Nicobar Islands")
plt.xticks(rotation=45)
plt.show()
```
